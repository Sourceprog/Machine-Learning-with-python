```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split

```

```python
dataset_cancer= sklearn.datasets.load_breast_cancer()

```

```python
print(dataset_cancer)
```

```python
data_frame = pd.DataFrame(dataset_cancer.data, columns= dataset_cancer.feature_names)
```

```python
data_frame.info()
```

```python
data_frame['label'] = dataset_cancer.target
```

```python
#information about the data_frame
data_frame.info()
```

```python
# to check with if we have any null values
data_frame.isnull().sum()
```

```python
#Analysis of statistical values of data_frame
data_frame.describe()
```

```python
# to check the target values counts
data_frame['label'].value_counts()
```

```python
#check the mean values based on labels
data_frame.groupby('label').mean()
```

```python
### Seperating the features and labels
X=data_frame.drop(columns='label',axis=1)
Y= data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
# splitting data into traning and testing data
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.3, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std= scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
# buiding NN
#import tensorflow, keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up layers on neural network
model=keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(50,activation='relu'),
                        keras.layers.Dense(2, activation='sigmoid')])
```

```python
#compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training neural network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=100)
```

```python
#
```

Visualizing accuracy and loss


```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('cpoch')

plt.legend(['training data','validation data'], loc= 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model accuracy')
plt.ylabel('loss')
plt.xlabel('cpoch')

plt.legend(['training data','validation data'], loc= 'upper right')
```

```python
#accuracy for test data
loss,accuracy= model.evaluate(X_test_std,Y_test)
print(accuracy)
```
